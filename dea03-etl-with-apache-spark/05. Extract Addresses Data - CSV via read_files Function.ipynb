<b>Extract Data From the Address Files 
1. Demostrate Limitations of CSV file format in SELECT statement 
2. Use read_files function to overcome the limitation 
3. Create Addresses view in the Bronze Layer

<b> 1. Demonstrate Limitations of CSV file format in SELECT statement

```python
%sql
SELECT * FROM csv.`/Volumes/gizmobox/landing/operational_data/Anddresses`
```

<b> 2. Use read_files function to overcome the limitations 

```python
%sql
SELECT * 
  FROM read_files('/Volumes/gizmobox/landing/operational_data/addresses',
                    format => 'csv', 
                    delimiter => '\t',
                    header => true);
```

<b> 3. Create Addresses view in the Bronze Layer

```python
%sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_addresses
AS
SELECT * 
  FROM read_files('/Volumes/gizmobox/landing/operational_data/addresses',
                    format => 'csv', 
                    delimiter => '\t',
                    header => true);
```

```python
%sql
SELECT * FROM gizmobox.bronze.v_addresses;
```
