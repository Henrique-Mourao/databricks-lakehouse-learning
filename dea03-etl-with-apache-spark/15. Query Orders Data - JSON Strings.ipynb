<b> Query Orders Data as JSON Strings 
1. Extract Top Level Column Values 
2. Extract Array elements
3. Extract Nested Column Values 
4. CAST Column Values to a Specific Data Type 

```python
%sql
SELECT *
FROM gizmobox.bronze.v_orders
```

<b> 1. Extract Top Level Object Values
<br>

Top Leve Object < column_name >:< extraction_path >


```python
%sql
SELECT value:order_id AS order_id,
value
FROM gizmobox.bronze.v_orders;
```

<b> 2. Extract Array elements
<br>

To extract array elements - use the indices in brackets < column_name >:< esxtraction_path > [index]

```python
%sql
SELECT value:items[0] AS items1,
       value:items[1] AS items2,
value
FROM gizmobox.bronze.v_orders;
```

<b> 3. Extract Nested Column Values
<br>
To extract nested fiekds, use dot notation < column_name >:< extraction_path > [index] . < leaf node >


```python
%sql
SELECT value:items[0].item_id AS item_id,
       value:items[0] AS item_1,
       value:items[1] AS item_2,
value
FROM gizmobox.bronze.v_orders;
```

<b>4. CAST Column Values to a Specific Data Type


```python
%sql
SELECT value:items[0].item_id::INTEGER item_id,
       value:items[0] AS item_1,
       value:items[1] AS item_2,
value
FROM gizmobox.bronze.v_orders;
```
