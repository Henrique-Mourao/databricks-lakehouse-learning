<b> User Defined Functions (UDF)
<br> 
> - User Defined Functions (UDF) in Spark are custom functions created by users to extend the capabilities of Spark SQL and PySpark
> - UDFs allow us to perform calculations or transfomations to apply business logic that are not possible with bult-in functions
> - You define the function once and use it accross multiple queries.
> - SQL UDFs are recommended over Python UDfs due to better optimization

#<b>Syntax </b>

---

<br>
<b> CREATE FUNCTION </b> catalog_name.schema_name.udf_name(param_name_data_type)
<br>
<b>RETURNS</b> return_type
<br>
<b>RETURN</b> expression

```python
%sql
CREATE OR REPLACE FUNCTION gizmobox.default.get_fullname(firstname STRING, surname STRING)
RETURNS STRING
RETURN CONCAT(initcap(firstname), ' ', initcap(surname));
```

```python
%sql
SELECT gizmobox.default.get_fullname('john', 'doe');
```

```python
%sql
DESC FUNCTION EXTENDED gizmobox.default.get_fullname;
```

```python
%sql
SELECT payment_id,
       customer_id,
       CAST(date_format(payment_date,'yyyy-MM-dd') AS DATE) AS payment_date,
       date_format(payment_date,'HH:mm:ss') AS payment_time,
       CASE payment_status
         WHEN 1 THEN 'Sucess'
         WHEN 2 THEN 'Pending'
         WHEN 3 THEN 'Cancelled'
         WHEN 4 THEN 'Failed'
       END AS payment_status,
       payment_method
  FROM gizmobox.bronze.payments;
```

```python
%sql
CREATE OR REPLACE FUNCTION gizmobox.default.get_payment_status(status INT)
RETURNS STRING
RETURN CASE
         WHEN status = 0 THEN 'Pending'
         WHEN status = 1 THEN 'Paid'
         WHEN status = 2 THEN 'Canceled'
         WHEN status = 3 THEN 'Failed'
         ELSE 'Unknown'
       END;
```

```python
%sql
SELECT payment_id,
       customer_id,
       CAST(date_format(payment_date,'yyyy-MM-dd') AS DATE) AS payment_date,
       date_format(payment_date,'HH:mm:ss') AS payment_time,
       gizmobox.default.get_payment_status(payment_status) AS payment_status,
       payment_method 
FROM gizmobox.bronze.payments;
```
