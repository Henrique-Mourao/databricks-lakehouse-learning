<b>Transform Payments Data

1. Extract date and time from payment_timestamp and create new columns payment_date and payment_time
2. Map payment_status to descriptive values (1‑Success, 2‑Pending, 3‑Cancelled, 4‑Failed)
3. Write the transformed data to the Silver schema

```python
%sql
SELECT payment_id,
       customer_id,
       payment_date,
       payment_status,
       payment_method
    FROM gizmobox.bronze.payments;
```

<b> 1.Extract Date and Time from payment_timestemp
<br>
> [Documention for date format function](https://learn.microsoft.com/en-us/azure/databricks/sql/language-manual/functions/date_format)

```python
%sql
SELECT payment_id,
       customer_id,
       CAST(date_format(payment_date, 'yyyy-MM-dd')AS DATE) AS payment_date,
       date_format(payment_date, 'HH:mm:ss') AS payment_time,
       payment_status,
       payment_method
    FROM gizmobox.bronze.payments;
```

<b>2. Map payment_status to contain descriptive valuesn
<br>
>(1-Success, 2-Peding, 3-Cancelled, 4-Failed)

```python
%sql
SELECT payment_id,
       customer_id,
       CAST(date_format(payment_date, 'yyyy-MM-dd')AS DATE) AS payment_date,
       date_format(payment_date, 'HH:mm:ss') AS payment_time,
       CASE payment_status
        WHEN 1 THEN 'SUCCESS'
        WHEN 2 THEN 'PEDING'
        WHEN 3 THEN 'CANCELLED'
        WHEN 4 THEN 'FAILED'
        ELSE 'UNKNOWN'
       END AS payment_status,
       payment_method
    FROM gizmobox.bronze.payments;
```

<b> 3. Write transformed data to the Silver schema 

```python
%sql
CREATE TABLE gizmobox.silver.payments
SELECT payment_id,
       customer_id,
       CAST(date_format(payment_date, 'yyyy-MM-dd')AS DATE) AS payment_date,
       date_format(payment_date, 'HH:mm:ss') AS payment_time,
       CASE payment_status
        WHEN 1 THEN 'SUCCESS'
        WHEN 2 THEN 'PEDING'
        WHEN 3 THEN 'CANCELLED'
        WHEN 4 THEN 'FAILED'
        ELSE 'UNKNOWN'
       END AS payment_status,
       payment_method
    FROM gizmobox.bronze.payments;
```

```python
%sql
SELECT * FROM gizmobox.silver.payments;
```
