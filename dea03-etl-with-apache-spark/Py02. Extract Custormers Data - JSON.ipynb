## <b> Extract Data From the Customers JSON File
1. Query Single file
2. Query List of Files Using wildcard Characters
3. Query all the files in a Folder


<b> 1. Query Silge JSON file

```python
%sql
SELECT * 
FROM json.`/Volumes/gizmobox/landing/operational_data/customers/customers_2024_10.json`
```

<b> 2. Query Multiple JSON files


```python
%sql
SELECT * 
FROM json.`/Volumes/gizmobox/landing/operational_data/customers/customers_2024_*.json`
```

<b> 3. Query all the files in a folder 

```python
%sql
SELECT * 
FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

<b> 4. Select File Metadata


```python
%sql
SELECT  _metadata.file_path AS file_path,
        *
      FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

<b> 5. Register Files in Unity Catalog using Views

```python
%sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_customers
AS
SELECT  _metadata.file_path AS file_path,
        *
      FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

```python
%sql
SELECT * FROM  gizmobox.bronze.v_customers;
```

<b> 6.Create Temporary View

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW tv_customers
AS
SELECT *,
       _metadata.file_path AS file_path
    FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

```python
%sql
SELECT * FROM  tv_customers;
```

<b> 7.Create Global Temporary View


```python
%sql
CREATE OR REPLACE TEMPORARY VIEW gtv_customers
AS 
SELECT *,
       _metadata.file_path AS file_path
    FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

```python
%sql
SELECT * FROM gtv_customers
```
