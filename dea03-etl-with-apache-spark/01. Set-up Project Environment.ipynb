<b>2.Create External Location 


```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS cloud_storage 
  URL 'abfss://<EMAIL>@<EMAIL>.dfs.core.windows.'
  WITH (STORAGE CREDENTIAL cloud_storage)
  COMMENT 'External Location For the Gizmobox Data Lakehouse'
```

<b>3. Criar Catalog

```python
%sql
CREATE CATALOG IF NOT EXISTS gizmobox
  MANAGED LOCATION 'abfss://henriquegmour4@gmail.com.dfs.core.windows.net/'
  COMMENT 'Gizmobox Data Lakehouse';
```

<b>4. Create Schemas</b>
1. Landing
2. Bronze
3. Silver
4. Gold



```python
%sql
SELECT current_catalog();
```

```python
%sql
USE CATALOG gizmobox;

CREATE SCHEMA IF NOT EXISTS landing
  MANAGED LOCATION 'abfss://henriquegmour4@gmail.com.dfs.core.windows.net/landing';
CREATE SCHEMA IF NOT EXISTS bronze
  MANAGED LOCATION 'abfss://henriquegmour4@gmail.com.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS silver
  MANAGED LOCATION 'abfss://henriquegmour4@gmail.com.dfs.core.windows.net/silver';
CREATE SCHEMA IF NOT EXISTS gold
  MANAGED LOCATION 'abfss://henriquegmour4@gmail.com.dfs.core.windows.net/gold';
```
