<b> Monthly Order Summary
1. total orders
2. total items bought
3. total amount spent

```python
%sql
SELECT * FROM gizmobox.silver.orders;
```

```python
%sql
SELECT date_format(transaction_timestamp, 'yyyy-MM-dd') order_month
FROM gizmobox.silver.orders;
```

```python
%sql
SELECT date_format(transaction_timestamp, 'yyyy-MM-dd') order_month,
       customer_id,
       COUNT(DISTINCT order_id) total_orders,
       SUM(quantity) total_items_bought,
       SUM  (price * quantity) total_amount
FROM gizmobox.silver.orders
GROUP BY order_month, customer_id;
```

```python
%sql
CREATE TABLE IF NOT EXISTS gizmobox.gold.order_sumary_monthly

SELECT date_format(transaction_timestamp, 'yyyy-MM-dd') order_month,
       customer_id,
       COUNT(DISTINCT order_id) total_orders,
       SUM(quantity) total_items_bought,
       SUM  (price * quantity) total_amount
FROM gizmobox.silver.orders
GROUP BY order_month, customer_id;
```
