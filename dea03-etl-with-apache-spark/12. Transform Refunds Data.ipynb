<b>Transform Refunds Data

1. Extract specific portion of the string from refund_reason using split function  
2. Extract specific portion of the string from refund_reason using regexp_extract function  
3. Extract date and time from the refund_timestamp  
4. Write transformed data to the Silver schema in hive metastore  

```python
%sql
SELECT  
    refund_id,  
    payment_id,  
    refund_timestamp,  
    refund_amount,  
    refund_reason  
FROM hive_metastore.bronze.refunds;
```

<b>1. Extract specific portion of the string from refund_reason using split function

[Documentation for split Function](link)

```python
%sql
SELECT  
    refund_id,  
    payment_id,  
    refund_timestamp,  
    refund_amount,  
    SPLIT(refund_reason, ':')  
FROM hive_metastore.bronze.refunds;
```

<b>2. Extract specific portion of the string from refund_reason using regexp_extract function
<br>
[Documentation for regexp_extract Function](link)

```python
%sql
SELECT
      refund_id,
      payment_id,
      CAST(date_format(refund_timestamp, 'yyyy-MM-dd') AS DATE) AS refund_date,
      date_format(refund_timestamp, 'HH:mm:ss') AS refund_time,
      refund_amount,
      regexp_extract(refund_reason, '([:]+):', 1) AS refund_reason,
      regexp_extract(refund_reason, '[:]+:(.*)$', 1) AS refund_source
FROM hive_metastore.bronze.refunds;
```

<b>3.Write transformed data to the Silver schema 

```python
%sql
CREATE SCHEMA hive_metastore.silver;
```

```python
%sql
CREATE TABLE hive_metastore.silver.refunds
AS
SELECT
      refund_id,
      payment_id,
      CAST(date_format(refund_timestamp, 'yyyy-MM-dd') AS DATE) AS refund_date,
      date_format(refund_timestamp, 'HH:mm:ss') AS refund_time,
      refund_amount,
      regexp_extract(refund_reason, '([:]+):', 1) AS refund_reason,
      regexp_extract(refund_reason, '[:]+:(.*)$', 1) AS refund_source
FROM hive_metastore.bronze.refunds;
```

```python
%sql 
SELECT * FROM hive_metastore.silver.refunds;
```
