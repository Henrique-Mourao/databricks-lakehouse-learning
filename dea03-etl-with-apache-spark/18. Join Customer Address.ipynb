<b> Join Customer and Address 
<br>

Join customer data with address data to create a customer_address table whitch contains the address of each customer on the same

```python
%sql
SELECT * FROM gizmobox.silver.customers;
```

```python
%sql
SELECT * FROM gizmobox.silver.addresses;
```

```python
%sql
CREATE TABLE IF NOT EXISTS gizmobox.gold.customers_address
SELECT c.customer_id,
       c.customer_name,
       c.email,
       c.date_of_birth,
       c.member_since,
       c.telephone,
       a.shipping_address_line_1,
       a.shipping_city,
       a.shipping_state,
       a.shipping_postcode,
       a.billing_address_line_1,
       a.billing_city,
       a.billing_state,
       a.billing_postcode
  FROM gizmobox.silver.customers c
  INNER JOIN gizmobox.silver.addresses a
          ON c.customer_id = a.customer_id 
```

```python
%sql
SELECT * FROM gizmobox.gold.customers_address
```
