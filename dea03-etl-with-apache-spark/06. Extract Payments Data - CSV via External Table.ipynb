<b> 2. Create External Table

```python
%sql
CREATE TABLE gizmobox.bronze.payments
(
  payment_id INT,
  customer_id INT,
  payment_date TIMESTAMP,
  payment_status STRING,
  payment_method STRING
)
USING DELTA;
```

```python
spark.read.text("dbfs:/Volumes/gizmobox/landing/operational_data/payments/payments_2024_10.csv").show(5, False)
```

```python
%sql
TRUNCATE TABLE gizmobox.bronze.payments;
```

```python
%sql
COPY INTO gizmobox.bronze.payments
FROM (
  SELECT
    CAST(_c0 AS INT) AS payment_id,
    CAST(_c1 AS INT) AS customer_id,
    CAST(_c2 AS TIMESTAMP) AS payment_date,
    CAST(_c3 AS STRING) AS payment_status,
    CAST(_c4 AS STRING) AS payment_method
  FROM 'dbfs:/Volumes/gizmobox/landing/operational_data/payments/'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
  'header'='false',
  'delimiter'=','
);
```

```python
%sql
SELECT * FROM gizmobox.bronze.payments;
```

```python
%sql
DESCRIBE EXTENDED gizmobox.bronze.payments;
```
