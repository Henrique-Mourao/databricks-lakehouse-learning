<b> Higher Order Functions
> - Higher Order Functions are functions that operate on complex data types such as arrays and maps
> - They allow you to pass functions as arguments (such as lambda)
> - They are extremely useful for manipulating arrray without exploding them

<b> Commonly used Higher Order Array Functions
- TRANSFORM
- FILTER
- EXISTS
- AGGREGATE

<b> Syntax
< function_name > (array_column, lambda_expression)
<br>

<i>lambda_expression:</i> element -> expression

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW order_items AS
SELECT * FROM 
VALUES 
  (1, array('smartphone', 'laptop', 'monitor')),
  (2, array('tablet', 'headphones', 'smartwatch')),
  (3, array('keyboard', 'mouse'))
  AS orders(order_id, items)
```

```python
%sql
SELECT * FROM order_items;
```

1. Convert all the item names to be UPPERCASE (TRANSFORM Function)


```python
%sql
SELECT order_id,
        TRANSFORM(items, x -> UPPER(x)) AS upper_items
  FROM order_items;
```

2. Filter only items that contains the string 'smart' (FILTER Function)
 

```python
%sql
SELECT order_id,
        TRANSFORM(items, x -> x LIKE '%smart%') AS smart_items
  FROM order_items;
```

<b> 3. Check to see whether the order includes any 'monitor' (EXISTS Function)

```python
%sql
SELECT order_id,
        EXISTS(items, x -> x = 'monitor') AS has_monitor
  FROM order_items;
```

Array with more than one object 


```python
%sql
CREATE OR REPLACE TEMP VIEW order_items AS
SELECT * FROM VALUES
  (1, array(
    named_struct('name', 'smartphone', 'price', 1000),
    named_struct('name', 'laptop', 'price', 2000),
    named_struct('name', 'monitor', 'price', 5000)
    )),
  (2, array(
    named_struct('name', 'tablet', 'price', 300),
    named_struct('name', 'headphones', 'price', 500), 
    named_struct('name', 'smartwatch', 'price', 1000)
    )),
  (3, array(
    named_struct('name', 'keyboard', 'price', 200),
    named_struct('name', 'mouse', 'price', 100)
    ))
  AS orders(order_id, items);
```

<b> 1. Convert all the item names to be UPPERCASE & Add 10% TAX to each item (TRANSFORM Function)

```python
%sql
SELECT order_id,
      TRANSFORM(items, x -> named_struct(
                                        'name', UPPER(x.name),
                                        'price', ROUND(x.price * 1.10, 2 )
                                        )) items_with_tax
      FROM order_items
```

<b> 2. Caculate the total order amount for each of the order (AGGREGATE Function)

```python
%sql
SELECT order_id,
       AGGREGATE(items, 0, (acc, x) -> acc + x.price, acc -> acc) AS total_order_price
    FROM order_items
```

# <b> Map Functions
<br>
A map is a collection of key_value pairs, like a dictionary {'laptop': 1200, 'phone': 800 }

## <b> Commonly used Higher Order Map Functions
- TRANSFORM_VALUES
- TRANSFORM_KEYS
- MAP_FILTER

<b> Syntax
--- 

< function_name > (map_column, lambda_expression)
<br>
<i>lambda expression:</i> (key, value) -> expression)

```python
%sql
CREATE OR REPLACE TEMP VIEW order_item_prices AS 
SELECT * FROM VALUES 
  (1, map('smartphone', 1000, 'laptop', 2000, 'monitor', 5000)),
  (2, map('tablet', 300, 'headphones', 500, 'smartwatch', 1000)),
  (3, map('keyboard', 200, 'mouse', 100))
  AS orders(order_id, item_prices);
```

```python
%sql
 SELECT * FROM order_item_prices
```

<b> 1. Convert all the item names to be UPPERCASE (TRANSFORM_KEYS Function)


```python
%sql
SELECT order_id,
        TRANSFORM_KEYS(item_prices, (item, price) -> UPPER (item)) AS items_upper_case
  FROM order_item_prices;
```

<b> 2. Appy 10% TAX to item prices (TRANSFORM_VALUES Function)

```python
%sql
SELECT order_id,
        TRANSFORM_VALUES(item_prices, (item, price) -> ROUND(price * 1.10, 2)) AS prices_with_tax
  FROM order_item_prices;
```

<b> 3. Filter only items with price above $500 (MAP_FILTER Functon) 

```python
%sql
SELECT order_id,
        MAP_FILTER(item_prices, (item, price) -> price > 500) AS premium_items
         FROM order_item_prices;
```
